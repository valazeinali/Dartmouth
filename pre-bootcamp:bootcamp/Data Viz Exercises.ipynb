```r
library(ggplot2)
library(dplyr)
install.packages("ggExtra")
```

```r
library(ggExtra)
options(warn=-1)
# classic plot :
p <- ggplot(df, aes(x=precipProbability, y=cloudCover, color=precipAccumulation, size=precipAccumulation, self_contained = TRUE)) +
      geom_point() +
      theme(legend.position="none")
 
# with marginal histogram
p1 <- ggMarginal(p, type="histogram")
 
# marginal density
p2 <- ggMarginal(p, type="density")
 
# marginal boxplot
p3 <- ggMarginal(p, type="boxplot")
```

```r
#library(scales)
#library(seq_gradient_pal)
ggplot(df, aes(x=precipProbability, y=precipAccumulation)) + 
    facet_wrap(~precipType)+
    geom_point(aes(color = cloudCover)) + 
    labs(x = "Probability of Precipitation",
         y = "Accumulation of Precipitation",
         subtitle = "Hanover, NH Weather Data",
         captions = "Colors are % of Cloud Coverage!"+
    scale_color_gradient(low = "green",
                            high = "blue")) 
    theme_bw() +
    theme(legend.position = "bottom",
          text = element_text(family = "serif"))
# found an error! There is no precipitation found on days labeled as "rain"
```
