```r
## Sequence of daily dates between "2018-12-31" and "2020-08-09", labeling them with seasons
df$season <- dip("2018-12-31", "2020-08-09",date.fmt = "%Y-%m-%d", out.type = "seq")
df$season = time2season(df$season)
```

# Q2: Using dplyr and %>%, subset the data to only include dates in 2020.

```r
options(warn=-1)
library(tidyverse)
```

```r
all_2020 <- df %>%
    filter(Date >= "2020-01-01") 
all_2020
```

# Q3: Filter the 2020 data to only include the partly cloudy days.



```r
cloudy <- all_2020 %>%
    filter(icon == "partly-cloudy-day") 
cloudy
```

# Q4: Using the filtered data, use "group_by" and "summarize", make a table which shows the median, mean, and sd of Temperature High by season. 

```r
cloudy %>%
  group_by(season) %>%
  summarize(mean_high = mean(temperatureHigh, na.rm = TRUE), med_high = median(temperatureHigh, na.rm = TRUE),
                                std_high = sd(temperatureHigh, na.rm = TRUE))
```

# Q5" Go back to the full 2020 day. Now use "group_by" and "summarize" to show the median, mean, and sd of precipitation accumulation and count of nice days by icon. 

```r
all_2020 %>%
  group_by(nice_or_not) %>%
  summarize(mean_high = mean(temperatureHigh, na.rm = TRUE), med_high = median(temperatureHigh, na.rm = TRUE),
                                std_high = sd(temperatureHigh, na.rm = TRUE))
```
