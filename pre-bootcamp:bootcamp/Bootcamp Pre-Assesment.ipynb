# Question 1
### Create a vector of numbers from 1 to 100, but replace multiples of 3 with “Fizz”, multiples of 5 with “Buzz”, and multiples of both with “FizzBuzz”. Save the vector as FB:

```r
fbnums <- 1:100
FB <- vector()

for (i in fbnums) {
  if (i %% 3 == 0 && i %% 5 == 0) {
    FB[i] <- "FizzBuzz"
  } else if (i %% 3 == 0) {
    FB[i] <- "Fizz"
  } else if (i %% 5 == 0) {
    FB[i] <- "Buzz"
  } else {
    FB[i] <- i
  }
}
```

```r
print(FB)
```

# Question 2
## Part a
### Create a vector with length 100 that randomly samples from the standard normal distribution. Call this V1:

```r
V1 <- vector(mode="character", length=100)


V1 = rnorm(100)

print(V1)
```

## Part b
### Create a vector with length 100 that randomly samples from the standard uniform distribution. Call this V2:

```r
V2 <- vector(mode="character", length=100)

V2 = runif(100, min = 0, max = 1)

print(V2)
```

## Part c
### Create a vector that results from multiplication of V1 by V2. Call this V3:

```r
V3 <- vector(mode="character", length=100)

V3 = V1*V2

print(V3)
```

## Part d
### Add noise (i.e., small random perturbations) to V3. Call this V4:

```r
V4 <- vector(mode="character", length=100)

V4 <- jitter(V3, factor = 1, amount = NULL)

print(V4)
```

## Part e
### Find the correlation between V1 and V3 & save it as Corr:

```r
Corr <- cor(V3, V1)
print(Corr)
```

## Part a

### Write a function (without using any built-in functions!) to calculate the mean of a vector of numbers, v:


```r
function(v) {
    s = sum(v)
    n = length(v)
    m = s/n
}
```

## Part b
### Repeat this for the median:

```r
function(v) {
    n <- length(v)
    s <- sort(v)
    ifelse(n%%2==1,s[(n+1)/2],mean(s[n/2+0:1]))
}
```

## Part c
### Repeat this for the maximum:

```r
function(v) {
    max(v)
}
```

## Part d
### Repeat this for the mode:


```r
function(v) {
    v <- sort(v)  
    u <- unique(v)
    y <- lapply(u, function(y) length(v[v==y]))
    u[which( unlist(y) == max(unlist(y)) )]
}
```
