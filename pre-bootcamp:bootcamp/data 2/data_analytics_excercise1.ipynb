```r
dat <- read.csv(file = "inflammation-01.csv", header = FALSE)
```

```r
plot_distribution <- function(df, threshold) {
  if (length(df) > threshold) {
    boxplot(df)
  } else {
    stripchart(df)
  }
}
```

```r
plot_distribution(dat[,10], threshold = 10)
```

```r
plot_distribution(dat[,10], threshold = 102)
```
