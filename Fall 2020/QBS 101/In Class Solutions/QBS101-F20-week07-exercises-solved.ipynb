# Exponentiate
Define a `exponentiate` function that takes in two numbers. The first is the base, and the second is the power to raise the base to. The function should work for positive and negative integer values of the 'power' and return the result of this operation. 
Caveat, you are not allowed to use the `**` operator or `pow()` function.

````
def exponentiate(x,n):
  exp=1
  if n<0:
    x = 1/x
    n = abs(n)

  for _ in range(n):
    exp*=x
  
  return exp
````

````
print(exponentiate(5,2))
print(exponentiate(5,-2))
````

# Better Age Calculation
Write a function `calculate_age` that takes a date of birth in the American date format `mm/dd/yyyy` as a string and the returns:
- age corresponding to that date, if the input is valid
- `none` if the input is invalid - implement 3 condition on the input to make your code reasonably defensive. Print a message before returning.

Call that function to test it.

````
from datetime import datetime as dt

def calculate_age(age_string):
  split=age_string.split('/')
  if len(split)!=3:
    print('Wrong input format')
    return

  m,d,y = [int(n) for n in split]
  if m < 1 or m > 12 or d > 31 or d < 1:
    print('Day or month invalid')
    return

  today = dt.today()
  age = today.year-y - (1 if today.month<m or (today.month==m and today.day<d) else 0)
  if(age<0):
    print('Date after today\'s date')
    return
    
  return age

print(calculate_age('12/13/1978'))
````

Now implement a loop that asks for the user input of a date and prints the age, until the user types `q` to quit the loop.

````
while True:
  date = input("DoB: ")
  if date=='q':
    break

  print(calculate_age(date))
````

# Prime
Write a function `is_prime` that takes in a single number and returns whether or not the number is prime.

Use a loop to test it up to a 100k.

````
def is_prime(x):
  for i in range(2,int(x/2)+1):
    if x%i==0:
      return False

  return True

for x in range(10**5):
  print(f'{x} is prime: {is_prime(x)}')
````
