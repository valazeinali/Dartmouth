## 2.2 - Games of chance - 15pts

Use the `random` package and its methods to:
- simulate flipping a single coin: sample one element from `heads` and `tails`
- simulate flipping 10 coins: sample elements from `heads` and `tails`, with replacement
- simulate casting 10 die: sample elements from integer values 1 through 6 with replacement
- simulate 100 roulette wheel spins: sample element form integers values between 0 to 36, boundaries included, with replacement (this is a European roulette, no "double-zero").

Print each simulation.

````python
# SOLUTION GOES HERE
import random
# part 1
print(f'One single fair coin flip result: {random.randint(1,2)}') # Single coin flip

# --------------------------------------------------------------------------------

# part 2
container_coin = [] #to store our results for each flip 
n = 10
i = 0
while(i<n): #10 simulations
    container_coin.append(random.randint(0,1)) # Appending results to our list 1 = heads, 0 =tails
    i+=1
print(f'Simulating 10 fair coin flips with replacement: {container_coin}')

# --------------------------------------------------------------------------------

# part 3
container_die = [] #to store our results for each die roll
n = 10
i = 0
while(i<n):
    container_die.append(random.randint(1,6)) # Appending results to our list 
    i+=1
print(f'Simulating 10 die rolls with replacement: {container_die}')

# --------------------------------------------------------------------------------

# part 4
container_roul = [] #to store our results for each roulette spin
n = 100
i = 0
while(i<n):
    container_roul.append(random.randint(0,36)) # Appending results to our list 
    i+=1
print(f'Simulating 100 roulette spins with replacement: {container_roul}')
````

## 4.2 - Patient Script - 30pts
Write a script that:
- asks for the patient's blood pressure in the "systolic/diastolic" format (eg. `115/75`, pronounced "115 over 75")
- prints the _blood pressure catergory_ according to the table
- make the code defensive:
  - blood pressure diastolic and systolic values are positive integers
  - the systolic value is always higher than the diastolic value
  - print an informative error message

````python
# SOLUTION GOES HERE
# make sure its a integer input
import sys

print("Hi! Please input your blood pressure in the following format: systolic/diastolic e.g, (115/75)")
bloodPressure = str(input("Whats your blood pressure? "))
temp = (bloodPressure.split('/')) #splitting the string on "/" into a list of strings

# converting the two strings in our list into integers, if the user did not properly type in our format throw an error
# with a special useful message (Defensive code)
try:
    sys1,dia = int(temp[0]),int(temp[1]) #first element becomes our systolic and second is diastolic and convert to int
except:
    sys.exit("You need to make sure you only have one / and two whole numbers!") #if input is a character throw this.
    
if(sys1 < 0):
    sys.exit("You cannot have a negative number for systolic or diastolic!")
elif(dia < 0):
    sys.exit("You cannot have a negative number for either systolic or diastolic!")
elif(dia > sys1): # edge case to check proper order
    sys.exit("Make sure you type your pressure in the correct order systolic/diastolic ")

#our chart
if((sys1 > 180) or (dia > 120)): # Hypertension 2
    print("Hypertensive crisis, contact doctor ASAP! ")
elif(sys1 < 120 and dia < 80):
    print("Your blood pressure is Normal")
elif((sys1 > 119 and sys1 < 130) and (dia < 80)): # Elevated
    print("Your blood pressure is elevated")
elif((sys1 > 129 and sys1 < 140) or (dia > 79 and dia < 90)): # Hypertension 1
    print("Hypertension stage 1! ")
elif((sys1 > 140) or (dia > 90)): # Hypertension 2
    print("Hypertension stage 2!! ")
````
