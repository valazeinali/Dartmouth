## 1.6 Cases vs. Recovery (3 points)
Let's assume that **on October 20th**, the rate of new cases begins to decrease at a predictible, steady rate of 10 cases per day. At the same time, recovered cases begin to increase at 50 cases per day. On which day will there be more individuals who have recovered than cases?


```python
# Your solution here
from datetime import date, timedelta

new_cases = 9746
recovered = 8258
for x in range(100):
    new_cases+=10
    recovered+=50
    if(recovered > new_cases):
        print(f'{x} days after October 20th, 2020, is when new cases {new_cases}, surpasses recovered {recovered}')
        date1 = date(2020, 10, 20)
        date2 = date1 + timedelta(days=x)
        print(f'This is the date of {date2}')
        break
```
