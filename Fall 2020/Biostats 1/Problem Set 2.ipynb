# 5. (Based on Rice 4.31) Let X be uniformly distributed on the interval [1,4]. Find E[1/X]. Is E[1/X] = 1/E[X]? Note: find E[X] using the definition of expectation, don’t just plug in the expectation of a U(a,b) RV.


<font color='red'>
    
\begin{align*}
E[\frac{1}{x}] = \int_{4}^{1} \frac{1}{x} f(x)dx = \int_{4}^{1} \frac{1}{x} \frac{1}{4-1}dx
\end{align*}   
\begin{align*}
= \frac{1}{3} \int_{4}^{1} \frac{1}{x}dx = \frac{1}{3} [ln|x| \Big|_1^4 ] = \frac{2}{3}ln(2)
\end{align*}       

\begin{align*}
\frac{1}{E[x]} = \frac{1}{\int_{4}^{1} \frac{1}{x} f(x)dx} = \frac{1}{\int_{4}^{1} \frac{1}{x} \frac{1}{4-1}dx}
\end{align*}       
\begin{align*}
= \frac{1}{\frac{1}{3}[\frac{x^2}{2} \Big|_1^4]} = \frac{5}{2}
\end{align*}
    
\begin{align*}
\therefore Since \, \frac{5}{2} \neq \frac{2}{3}ln(2), \; E[\frac{1}{x}] \neq \frac{1}{E[x]}
\end{align*}

# 6. (Based on Rice 4.49) Two independent measurements, X and Y, are taken of a quantity μ. E[X] = E[Y] = μ, but σx and σy are unequal. The two measurements are combined by means of a weighted average to give:
\begin{align*}
Z = αX + (1 − α)Y
\end{align*}

## where α is a scalar and 0 ≤ α ≤ 1.


* A Show that E[Z] = μ.
* B If X and Y are not independent, what is E[Z]?
* C What is Var(Z)? Does this result hold if X and Y are not independent?
* D Find α in terms of σX and σY to minimize Var(Z).
* E Under what circumstances is it better (i.e., minimizing variance) to use the average (X + Y )/2 than either X or Y alone?


<font color='red'>
a)
\begin{align*}
E[Z] = E[\alpha X + ((1-\alpha)Y)
\end{align*}    
\begin{align*}
= \alpha E[X] + (1-\alpha)E[Y]
\end{align*}        
\begin{align*}
= \alpha \mu +\mu - \alpha \mu = \mu
\end{align*}
    

    
b)
Same as part a.
\begin{align*}
E[Z] = E[\alpha X + ((1-\alpha)Y)
\end{align*}    
\begin{align*}

\end{align*}        
\begin{align*}
= \alpha \mu +\mu - \alpha \mu = \mu
\end{align*} 
    
    
    
c)
\begin{align*}
Var(Z)=Var(\alpha X + ((1-\alpha)Y)
\end{align*}    
\begin{align*}
= \alpha^2 \sigma^2_X + (1-\alpha)^2 \sigma^2_Y = f(a)
\end{align*}      

    
    
d)
\begin{align*}
{f}'(\alpha) = 2\alpha \sigma^2_X - (1-\alpha)\sigma^2_Y
\end{align*}    
\begin{align*}

\end{align*}
    
\begin{align*}

\begin{align*}
\therefore  2\alpha \sigma^2_X = 2(1-\alpha)\sigma^2_Y 
\Rightarrow 2\sigma^2_X + 2\sigma^2_Y = 2\alpha \sigma^2_Y
